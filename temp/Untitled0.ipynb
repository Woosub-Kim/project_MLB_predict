```
import lxml
import pandas as pd
year = [str(i) for i in range(2001,2021)]
month = ['04','05','06','07','08','09']
day = [str(i) if i>= 10 else '0'+str(i) for i in range(1,32)]
score_list = []

for y in year:
    for m in month:
        print(y,m)
        for d in day:
            date = y+m+d
            url = 'https://www.espn.com/mlb/schedule/_/date/'+date
            try:
                SCORE = pd.read_html(url)[0][['matchup', 'Unnamed: 1', 'result']]
                SCORE.columns = ['AWAY', 'HOME', 'RESULT']
                SCORE['DATE'] = date
                score_list.append(SCORE)
            except:
                continue

SCORES = pd.concat(score_list)
SCORES.to_csv('MLB_scores.csv')
```

```
url = 'http://www.espn.com/mlb/stats/batting/_/year/{}/count/{}/qualified/false'
import pandas as pd
year = [i for i in range(2000,2020)]
count = [i for i in range(1,1201,40)]
for y in year:
  print(y)
  foo = pd.DataFrame()
  for c in count:
    try:
      df = pd.read_html(url.format(y,c), header=1)[0]
      foo = pd.concat([foo,df])
    except:
      print(c)
  foo.to_csv('./war/WAR{}.csv'.format(y))
```

```
from SearchList import getSearchList
import numpy as np
import pandas as pd
year = [2011,2012,2013,2014,2015,2016,2017,2018,2019]
month = ['04','05','06','07','08','09']
day = [i for i in range(1,32)]
foo = pd.DataFrame()
for y in year:
  print(y)
  for m in month:
    for d in day:
      date = '{}-{}-{}'.format(y,m,d)
      L =  getSearchList('https://www.baseballpress.com/lineups/{}'.format(date),'div.lineup-card')
      temp = []
      
      for l in L: #div.lineup-card중에서
        # 라인업 비공개 경기 제거하기
        if 'No Lineup Released' in str(l):
          continue
        elif 'TBD' in str(l):
          continue
        #정상경기의 a태그
        lineup = [i.text.strip() for i in l.findAll('a')]
        #팀명 2개 + 선발 10*2
        if len(lineup)==22:
          temp.append(lineup)
          
      temp = np.array(temp)
      df = pd.DataFrame(temp)
      df['date'] = date
      foo = pd.concat([foo,df])
      
      
foo.to_csv('line_up2.csv')
```
